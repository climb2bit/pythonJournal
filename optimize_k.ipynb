```python
## Objective
## - use numpy to find optimal values for a and b in y = ax + b
## - use gradient descent to minimize the mean squared error

import numpy as np
import matplotlib.pyplot as plt

# Generate synthetic data, F = ma
F = np.array([9.8 * i * 50 / 1000 for i in [0, 1, 2, 3, 4]])  # Force in Newtons

# length of spring in centimeters
x = np.array([3.0, 3.4, 3.7, 3.9, 4])
```

```python
# fit a linear model F = kx + b
# explain what np.vstack and np.linalg.lstsq do
# np.vstack([x, np.ones(len(x))]).T creates a matrix A where each row is [x_i, 1]
# This represents the system of equations Ax = b where x is [k, b]
# given F = kx + b, we can rewrite it as F = A * [k, b]
# for example, if x = [3.0, 3.4], then A = [[3.0, 1], [3.4, 1]]
# 
# np.linalg.lstsq(A, F, rcond=None) solves for [k, b] that minimizes the squared error ||Ax - F||^2
# for example, if A = [[3.0, 1], [3.4, 1]] and F = [0.0294, 0.03332], it finds k and b such that
# the line F = kx + b best fits the data points (3.0, 0.0294) and (3.4, 0.03332)
A = np.vstack([x, np.ones(len(x))]).T
k, b = np.linalg.lstsq(A, F, rcond=None)[0]
print(f"Optimal spring constant k: {k:.4f} N/cm, intercept b: {b:.4f} N")
```

```python
# visualize the data, with fitted line
plt.scatter(x, F, marker='X', color='blue')
plt.plot(x, k * x + b, 'r', label='Fitted line')
plt.xlabel('Spring Length (cm)')
```
